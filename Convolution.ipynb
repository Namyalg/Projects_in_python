```
import numpy as np
import cv2

# Load an color image in grayscale
img = cv2.imread('/content/Lion.JPG',0)


img = cv2.resize(img, (300, 360))
cv2_imshow(img)
```

```
sharpen = np.array([0.11,0.11,0.11,0.11,0.11,0.11,0.11,0.11,0.11]).reshape(3,3)
print(sharpen)
```

```
pixels = []
for i in range(360):
  for j in range(300):
    pixels.append(img[i,j])

pixels = np.array(pixels).reshape(360,300)
pixels = np.insert(pixels , [0,360] , np.zeros(len(pixels[0])) , axis = 0)
pixels = np.insert(pixels , [0,300] , np.zeros((len(pixels[:, 0]) ,1)) , axis = 1)
print(pixels.shape)

reduced = []
for i in range(1,360):
  for j in range(1,300):
    d = pixels[i:i+3 , j:j+3]
    m = np.multiply(d,sharpen)
    reduced.append(sum(sum(m)))
  
reduced = np.array(reduced).reshape(359,299)
```

```
import numpy as np
from PIL import Image
array = (reduced).astype(np.uint8)
img = Image.fromarray(array)
img.save('blur.jpg')
```
