```
        upto20 = ['',"One" ,"Two" ,"Three" ,"Four" , "Five" ,"Six" , "Seven" ,  "Eight" ,"Nine" ,"Ten" , "Eleven" ,  "Twelve" , "Thirteen" , "Fourteen",  "Fifteen" ,  "Sixteen" ,"Seventeen" ,  "Eighteen" , "Nineteen"]
        tens = ["" ,"" ,"Twenty" ,"Thirty" ,"Forty" ,"Fifty" , "Sixty" , "Seventy" , "Eighty" ,"Ninety"]

        place_in_indian = ['Crore' , 'Lakh' , 'Thousand' ,'']
        place_in_western = ['Trillion','Billion' , 'Million' ,'Thousand' ,'']

        split_by_in_indian = 2
        split_by_in_western = 3

        def twodigits(n : int) -> str:
          if n == 0:
            return("Zero")
          elif n < len(upto20):
            return upto20[n]
          return tens[n//10] + ' ' + upto20[n%10]

        def threedigits(n : int) -> str:
          if n < 100:
            return twodigits(n)
          elif n % 100 == 0:
            return upto20[n//100] + H100
          return upto20[n//100] + ' ' + H100  + ' and ' + twodigits(n%100)

        def split_into_denominations(number : int):
          without_hundreds_tens_ones = str(number)[:-3]
          hundreds_ten_ones = str(number)[-3:]
          return(without_hundreds_tens_ones , hundreds_ten_ones)

        def convert_to_number_names(out : [] , place_values : []) -> str:
          store = ' '
          index = 0
          for i in out:
            if i == 0:
              continue
            store += threedigits(i) + ' ' + place_values[index] + ' '
            index += 1
          return(store)

        def modify_indian(without_h_t_o : int) -> int:
          if len(without_h_t_o) % 2 == 1:
            return('0' + without_h_t_o)
          return(without_h_t_o)

        def modify_western(without_h_t_o : int) ->int:
          if len(without_h_t_o) % 3 == 1:
            return('00' + without_h_t_o)
          elif len(without_h_t_o) % 3 == 2:
            return('0' + without_h_t_o)
          return(without_h_t_o)

        def split_without_h_t_o(without_h_t_o , h_t_o , split_by : int) -> []:
          out = [int(without_h_t_o[i:i+split_by]) for i in range(0 ,len(without_h_t_o) , split_by)]
          out.append(int(h_t_o))
          return(out)

        def get_the_place_values(place_in : [] , out : []) -> []:
          return(place_in[len(place_in) - len(out) : ])

        def indian(number : int) ->str:
          without_h_t_o , h_t_o = split_into_denominations(number)
          out = split_without_h_t_o(modify_indian(without_h_t_o) , h_t_o , split_by_in_indian)
          return(convert_to_number_names(out , get_the_place_values(place_in_indian , out)) + ' ' )


        def western(number : int) -> str:
          without_h_t_o , h_t_o = split_into_denominations(number)
          out = split_without_h_t_o(modify_western(without_h_t_o) , h_t_o , split_by_in_western)
          return(convert_to_number_names(out , get_the_place_values(place_in_western , out)) + ' ')

def number_to_words(number: int):
          in_ind_and_western = "IN THE INDIAN SYSTEM : " + indian(number).lower() + '\n' + "IN THE WESTERN SYSTEM : " + western(number).lower()
          return(in_ind_and_western) 


print(number_to_words(142919))
```
